```python
def get_SFRDzZ(redshifts, metals = [], min_logZ_COMPAS = np.log(1e-4), max_logZ_COMPAS = np.log(0.03),
               metal_params = [], SFR_Params = [], min_logZ=-12.0, max_logZ=0.0, step_logZ =0.01): 
    """
    """
    mu_0, mu_z, omega_0, omega_z, alpha = metal_params
    a, b, c, d                          = SFR_Params
    # metallicity distribution
    dPdlogZ, metallicities, step_logZ, p_draw_metallicity = \
                    Z_SFRD.skew_metallicity_distribution(redshifts, mu_0=mu_0, mu_z=mu_z,alpha = alpha, 
                                                  omega_0=omega_0, omega_z =omega_z, min_logZ  =-12.0, max_logZ  =0.0, step_logZ = 0.01,
                                                  metals = metals ) #np.logspace(-5,1, num=1000)
    # SFR
    sfr        = Z_SFRD.Madau_Dickinson2014(redshifts, a=a, b=b, c=c,  d=d) # Msun year-1 Mpc-3 
    # Combine it into a SFRD
    model_SFRD = (sfr*(dPdlogZ  ).T ).value #step_logZ
    return model_SFRD, metallicities
```

## Load the TNG 100 data

This starformation rate comes from the gas particles in TNG: StarFormationRate = "Instantaneous star formation rate of this gas cell."

See also: 
https://www.tng-project.org/data/docs/specifications/#parttype0

This is why we have to convert the rate to a density by dividing by the co-moving box size. 




```python
##############################################################################
# Load TNG100 data
with h5py.File(paths.data / "SFRMetallicityFromGasTNG100.hdf5", "r") as file:
    MetalBins            = file["MetalBins"][:] #60
    TNG100_center_Zbin   = (MetalBins[:-1] + MetalBins[1:])/2.
    TNG100_Lookbacktimes = file["Lookbacktimes"][:] #100
    BoxSfr               = file["Sfr"][:]
# Convert SFR from sfr/box to sfr cMpc-3
littleh = 0.6774
Rbox    = 75/littleh
TNG100_cosmic_SFR = BoxSfr / Rbox**3 *u.Mpc**-3
# TNG100_cosmic_SFR       = TNG100_cosmic_SFR.T

"""
 We are going to convert SFR to total stellar mass formed
"""
# Take the SFRD at the center of each bin in lookback time
TNG100_centered_SFR       = (TNG100_cosmic_SFR[:-1,:] + TNG100_cosmic_SFR[1:,:])/2.
# Multiply this by the diff in lookback time to get the total stellar mass formed in each lookback time bin
TNG100_Lookbacktimes_yr   = TNG100_Lookbacktimes *u.Gyr.to(u.yr)
# This gives us the stellar mass formed at t, per Mpc^-3 per d logZ (it's still 2D)
dlogZ    = np.diff(np.log10(MetalBins))[0]
totDlogZ = max(np.log10(MetalBins))  - min(np.log10(MetalBins))
print('dlogZ',dlogZ, 'totDlogZ', totDlogZ)
TNG100_stellarM_formed_dZ = (TNG100_centered_SFR * abs(np.diff(TNG100_Lookbacktimes_yr))[:, np.newaxis]) 

print('TNG100_stellarM_formed_dZ', np.shape(TNG100_stellarM_formed_dZ) )

"""
 Now to have the the total stellar mass formed up to some redshift, we need the z 
""" 

# First we need to convert observed lookback times to observed redshifts 
# (lookback times are decreasing, so also TNG100_redshift is decreasing)
TNG100_redshifts         = [z_at_value(cosmo.lookback_time,t*u.Gyr) for t in TNG100_Lookbacktimes[:-1]]
TNG100_redshifts.insert(len(TNG100_redshifts),0) # put redshift zero at the end (because cosmo.lookback_time cant handle 0)
TNG100_redshifts         = np.array(TNG100_redshifts)

# And take the center of the redshift bins to match the shape of TNG100_stellarM_formed_dZ
TNG100_center_redshifts  = (TNG100_redshifts[:-1] + TNG100_redshifts[1:])/2.
print('TNG100_center_redshifts',np.shape(TNG100_center_redshifts))

# # Now we can make a redshift bool
# redshift_points  = [0.5, 3., 10]

# redsh_bools  = [TNG100_center_redshifts < 10, TNG100_center_redshifts < 3.0, TNG100_center_redshifts < 0.5 ]

# # And the metallicity boundaries for 'low' and 'high' metallicity
# low_metals   = TNG100_center_Zbin < 0.014/10.
# high_metals  = TNG100_center_Zbin > 0.014

# for i, redsh_bool in enumerate(redsh_bools):
#     print('redshift', redshift_points[i])
#     # select all points with z < x
#     TNG100_stellarM_formed_dZ_up_toz = TNG100_stellarM_formed_dZ[redsh_bool, :]
# #     print('z_bool',redsh_bool)
#     print('total low Z Msun formed up to reds', np.sum(TNG100_stellarM_formed_dZ_up_toz[:, low_metals]) )
#     print('total high Z Msun formed up to reds', np.sum(TNG100_stellarM_formed_dZ_up_toz[:, high_metals]) )
#     print('total Msun formed up to reds', np.sum(TNG100_stellarM_formed_dZ_up_toz[:, :]) )



```
